```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load image from file
img = cv2.imread('./photo.jpeg')

# Show the image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
# Define the region of interest (ROI)
x, y, w, h = 250, 300, 550, 600 # top-left coordinates, width, and height
img_cropped = img[y:y+h, x:x+w]

# Show the image
plt.imshow(img_cropped)
plt.show()
```

```python
# Convert image to grayscale
img_gray = cv2.cvtColor(img_cropped, cv2.COLOR_RGB2GRAY)

# Show the image
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# Resize image
w, h = 1000, 1000
img_resized = cv2.resize(img_cropped, (w, h))

# Show the image
plt.imshow(img_resized)
plt.show()
```

```python
# Invert image
img_neg = cv2.bitwise_not(img_resized)

# Show the image
plt.imshow(img_neg)
plt.show()
```

```python
cv2.imwrite('./photo_neg.jpeg', cv2.cvtColor(img_neg, cv2.COLOR_RGB2BGR))
```
